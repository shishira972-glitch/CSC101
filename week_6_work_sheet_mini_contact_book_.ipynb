```python

contacts = {
    "jenish": {"phone": "0400 123 456", "email": "jenishe@example.com"},
    "ram": {"phone": "0411 111 222", "email": "ram@example.com"},
    "sadie": {"phone": "0422 555 333", "email": "sadie@example.com"},
    "Shishir": {"phone": "0433 777 888", "email": "shishir@example.com"}
}
for name, info in contacts.items():
    print(name)
    print("  Phone:", info["phone"])
    print("  Email:", info["email"])
```
