## Practice Exercise 1.1: Sales Performance by Region

::: {.callout-tip}
## Your AI Collaboration Challenge

**Business Question**: "Which region is performing best, and what's the average sale size by region?"

**Your Task**: Practice prompting AI with this business context. Try creating a prompt that asks for:
1. Total sales by region
2. Average sale amount by region  
3. Number of transactions by region
4. Results formatted for a manager presentation

**Write your prompt below, then see what AI generates:**
:::

**My Prompt for 1.1:**
```
Consider that I am a Business Analyst working for an office supply company and I have a data set file sales_data.csv with  with columns: Date: Sale date,  Product: Product name,  Quantity: Units sold , Unit_Price: Price per unit (with $ signs),  Total_Sale: Total transaction value , Sales_Rep: Employee who made the sale , Region: Geographic region .Using this  dataset, help me wrtie a program in pandas 

The program should:
1. Calculate total sales by region
2. Calculate average sale amount by region
3. Calculate the number of transactions by region
4. Format the results in a professional table suitable for a manager presentation
5. Include a short business quenstion answer identifying the best-performing region and what's the average sale size by region
Make sure the code is clean, beginner-friendly, and well-commented.
```

**AI Response:**

````python
import pandas as pd

# Load the dataset
sales_df = pd.read_csv("sales_data.csv")

# Remove '$' sign from Total_Sale column
# Convert the column into numeric format
sales_df["Total_Sale"] = (
    sales_df["Total_Sale"]
    .replace(r'[\$,]', '', regex=True)
    .astype(float)
)

# Group the data by Region and calculate:
# - Total Sales
# - Average Sale Amount
# - Number of Transactions
region_summary = sales_df.groupby("Region").agg(
    Total_Sales=("Total_Sale", "sum"),
    Average_Sale_Size=("Total_Sale", "mean"),
    Number_of_Transactions=("Total_Sale", "count")
).reset_index()

# Round average sales to 2 decimal places
region_summary["Average_Sale_Size"] = region_summary["Average_Sale_Size"].round(2)

# Format Total Sales column for presentation
region_summary["Total_Sales"] = region_summary["Total_Sales"].map('${:,.2f}'.format)

# Format Average Sale Size column for presentation
region_summary["Average_Sale_Size"] = region_summary["Average_Sale_Size"].map('${:,.2f}'.format)

# Display the final professional table
print("Sales Performance by Region")
print("-" * 50)
print(region_summary)

# Find the best-performing region based on highest total sales
best_region = sales_df.groupby("Region")["Total_Sale"].sum().idxmax()

# Display business insight
print("\nBusiness Insight:")
print(f"The best-performing region is {best_region} because it generated the highest total sales.")

print("\nAverage Sale Size by Region:")
for index, row in region_summary.iterrows():
    print(f"{row['Region']}: {row['Average_Sale_Size']}")
````

## Practice Exercise 1.2: Top Performing Products

::: {.callout-tip}
## Business Scenario

Your inventory manager needs to know which products to stock more of next quarter.

**Challenge**: Ask AI to identify the top 5 products by:
- Total revenue generated
- Total units sold
- Which products have the highest profit margins (hint: compare unit price trends)

**Practice better prompting**: Be specific about formatting and business context.
:::

**My Prompt for 1.2:**
```
Write a simple, beginner-friendly Python program using pandas to analyze a dataset called sales_data.csv. The goal is to help a manager decide which products should be stocked more in the next quarter by identifying the top 5 products based on total revenue, total units sold, and highest profit potential (using average unit price as an indicator).The output should be printed in a clean office-style format with clear headings, separators, and a few light emojis for readability and that extra look. Finally, include a short business insight section with simple recommendations for inventory management, keeping the explanation clear and beginner-friendly.
```

**AI Response:**

````python
import pandas as pd

# ==============================
# Load Dataset
# ==============================
sales_df = pd.read_csv("sales_data.csv")

# ==============================
#  Clean Data
# ==============================
sales_df["Unit_Price"] = sales_df["Unit_Price"].replace(r'[\$,]', '', regex=True).astype(float)
sales_df["Total_Sale"] = sales_df["Total_Sale"].replace(r'[\$,]', '', regex=True).astype(float)

# ==============================
# TOP 5 PRODUCTS BY REVENUE
# ==============================
top_revenue = (
    sales_df.groupby("Product")["Total_Sale"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# Add dollar sign
top_revenue["Total_Sale"] = top_revenue["Total_Sale"].map('${:,.2f}'.format)

# ==============================
# TOP 5 PRODUCTS BY UNITS SOLD
# ==============================
top_units = (
    sales_df.groupby("Product")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# ==============================
#  TOP 5 BY PROFIT POTENTIAL (Unit Price Trend)
# ==============================
profit_margin = (
    sales_df.groupby("Product")["Unit_Price"]
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

#  Add dollar sign
profit_margin["Unit_Price"] = profit_margin["Unit_Price"].map('${:,.2f}'.format)

print("\n TOP 5 PRODUCTS BY TOTAL REVENUE")
print("-" * 70)
print(top_revenue.to_string(index=False))
print("-" * 70)

print("\n TOP 5 PRODUCTS BY TOTAL UNITS SOLD")
print("-" * 70)
print(top_units.to_string(index=False))
print("-" * 70)

print("\n TOP 5 PRODUCTS BY PROFIT POTENTIAL")
print("-" * 70)
print(profit_margin.to_string(index=False))
print("-" * 70)
````

**Business Insight Questions :**

- The best performing region is NSW based on total sales.
- High-value technology products generate the most revenue.
- Lower-cost office products sell in larger quantities.
- NSW contributes the highest company revenue.
- The business should maintain stock for high-demand products.

## Practice Exercise 2.2: Budget Category Analysis

::: {.callout-tip}
## Personal Finance Scenario

Someone wants to set a realistic monthly budget based on their spending history.

**Challenge**: Use AI to help:
1. Calculate average monthly spending by category
2. Identify categories with high variation (inconsistent spending)
3. Suggest realistic budget amounts for each category
4. Find categories where spending might be reduced

**Advanced**: Ask AI to explain the statistical concepts it uses.
:::

**Your Budget Analysis:**

````python
import numpy as np
import pandas as pd

# Load dataset
df = pd.read_csv("transactions.csv")

# Clean column names
df.columns = df.columns.str.strip()

if "amount" in df.columns:
    df.rename(columns={"amount": "Amount"}, inplace=True)

if "category" in df.columns:
    df.rename(columns={"category": "Category"}, inplace=True)

# Clean Amount column
df["Amount"] = (
    df["Amount"]
    .astype(str)
    .str.replace("[$,]", "", regex=True)
)

df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")
df = df.dropna(subset=["Amount", "Category"])

# =========================
# ANALYSIS
# =========================
avg = df.groupby("Category")["Amount"].mean()
std = df.groupby("Category")["Amount"].std()
total = df.groupby("Category")["Amount"].sum()

budget_summary = df.groupby("Category")["Amount"].agg(['sum', 'mean', 'std'])
budget_summary["std"] = budget_summary["std"].fillna(0)
budget_summary["Suggested_Budget"] = budget_summary["mean"] * 1.10

# =========================
# CLEAN OUTPUT (NO dtype/name lines)
# =========================
print("\nAVERAGE SPENDING BY CATEGORY")
print(avg.to_string())

print("\nSPENDING VARIATION BY CATEGORY")
print(std.to_string())

print("\nTOTAL SPENDING BY CATEGORY")
print(total.to_string())

print("\nBUDGET SUMMARY")
print(budget_summary.to_string())

print("\nSUGGESTED MONTHLY BUDGET")
print(budget_summary[["Suggested_Budget"]].to_string())

# =========================
# INSIGHTS
# =========================
print("\nBUDGET INSIGHTS")
print("- Focus on categories with high average spending.")
print("- High variation means unstable spending patterns.")
print("- Suggested budget is based on past behavior with a small buffer.")
print("- Reduce non-essential spending categories if needed.")
````

## Scenario 2: Personal Finance Coaching

::: {.callout-note}
## Business Context

You're building a personal finance app feature that provides spending advice.

**Challenge**: Use AI to create code that:
1. Identifies spending categories that are above typical percentages
2. Suggests specific areas for potential savings
3. Provides encouraging feedback for good spending habits
4. Formats advice in a user-friendly way
:::

**Your Coaching Algorithm:**

````python
#using transaction.cvs dataset 

import pandas as pd

# Load data
df = pd.read_csv("transactions.csv")

# Clean Amount safely
df["Amount"] = (
    df["Amount"]
    .astype(str)
    .str.replace("[$,]", "", regex=True)
)

df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

# Remove rows with missing values (important fix)
df = df.dropna(subset=["Amount"])

# Convert date safely
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Date"])

total_spent = df["Amount"].sum()

category_spending = df.groupby("Category")["Amount"].sum().reset_index()
category_spending["Percentage"] = (category_spending["Amount"] / total_spent) * 100

# Benchmarks
benchmarks = {
    "Rent": 30,
    "Food": 15,
    "Transport": 10,
    "Utilities": 10,
    "Entertainment": 10,
    "Shopping": 10
}

print(f"Total Spending: ${total_spent:.2f}\n")

for _, row in category_spending.iterrows():
    cat = row["Category"]
    amt = row["Amount"]
    pct = row["Percentage"]
    limit = benchmarks.get(cat, 10)

    print(f"Category : {cat}")
    print(f"Spent    : ${amt:.2f}")
    print(f"Share    : {pct:.1f}% (Limit: {limit}%)")

    if pct > limit:
        print("Status   : Overspending")
        print("Advice   : Reduce spending in this category.")
    else:
        print("Status   : Healthy ")
        print("Advice   : Good financial control.")
````
