## Problem 1: Tip Calculator
### Step 1 — Understand the Problem

This program helps restaurant staff work out tips quickly. You enter the bill amount and the tip percentage you want, and the program calculates the tip in dollars and the total amount to charge the customer.
### Step 2 — Identify Inputs and Outputs

**Inputs:**
- Bill amount — a decimal number (e.g., $45.80)
- Tip percentage — a number (e.g., 15 for 15%)

**Outputs:**
- Tip amount — shown in dollars and cents (e.g., $6.87)
- Total amount — shown in dollars and cents (e.g., $52.67)
### Step 3 — Work It By Hand

**Example:** Bill = $45.80, Tip = 15%

- Convert 15% to a decimal: 15 ÷ 100 = 0.15
- Multiply bill by decimal: $45.80 × 0.15 = $6.87
- Add tip to bill: $45.80 + $6.87 = $52.67

**What I'd tell the staff member:**  
"The tip is $6.87, so the total to charge is $52.67."
## Problem 2: Late Fee Calculator
### Step 1 — Understand the Problem

This program helps a librarian calculate fines for overdue books. You enter how many days late a book is, and the program works out the fine at $0.50 per day. If the fine goes over $10.00, it shows a message saying the maximum has been reached instead.
### Step 2 — Identify Inputs and Outputs

**Inputs:**
- Days overdue — a whole number (e.g., 5, 25, 0)

**Outputs:**
- The fine amount in dollars and cents (e.g., $2.50)
- OR the message "Maximum fine reached" if over $10.00
- If 0 days late, the fine is $0.00
### Step 3 — Work It By Hand

**Case A: 5 days late**

- 5 × $0.50 = $2.50
- $2.50 is under $10.00, so no cap
- **Fine: $2.50**

---

**Case B: 25 days late**

- 25 × $0.50 = $12.50
- $12.50 is over $10.00, so the cap applies
- **Fine: Maximum fine reached**

*Why this is different:* The fine would be $12.50 normally, but the library caps it at $10.00. So the program shows the message instead of a dollar amount.

---

**Case C: 0 days late**

- 0 × $0.50 = $0.00
- No cap needed
- **Fine: $0.00**
