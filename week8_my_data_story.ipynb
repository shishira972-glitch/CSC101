```python
# File name
filename = "my_data.txt"

# Read file
with open(filename, "r") as file:
    lines = file.readlines()

print("=== MY DATA STORY ===")

# Show content
for line in lines:
    print(line.strip())

# Simple statistics
print("\n=== STATS ===")
print("Total lines:", len(lines))

# Count words
words = 0
for line in lines:
    words += len(line.split())

print("Total words:", words)
```
