```python
filename = "my_data.txt"

with open(filename, "r") as file:
    lines = file.readlines()

print("=== MY DATA STORY ===")

for line in lines:
    print(line.strip())

print("\n=== STATS ===")
print("Total lines:", len(lines))

words = 0
for line in lines:
    words += len(line.split())

print("Total words:", words)
```
